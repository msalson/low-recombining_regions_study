## Jupyter Notebook to obtain statistics concerning alignments of optical maps to a reference


#### XMAP and CMAP files

```python
fileXmap = "./103sr8_CO.xmap"

fileCmapRef = "./103sr8_CO_r.cmap"

fileCmapQuery = "./103sr8_CO_q.cmap"
```

```python
def file_to_df(fileXmap) :
    liste=[]
    with open(fileXmap, "r") as f1 :
        cpt_header=int(0)
        for line in f1 :
            if line[0]=="#" : 
                cpt_header+=1
                liste.append(int(cpt_header)-1)
            else :
                break
    del liste[-2]
    return liste
```

```python
import pandas as pa
from matplotlib import pyplot as plt
import numpy as np

import os
import re


liste = file_to_df(fileXmap)
xmap_df = pa.read_csv(fileXmap, sep='\t', skiprows=liste) 

liste2 = file_to_df(fileCmapRef)
cmap_df_ref = pa.read_csv(fileCmapRef, sep='\t', skiprows=liste2) 

liste3 = file_to_df(fileCmapQuery)
cmap_df_query = pa.read_csv(fileCmapQuery, sep='\t', skiprows=liste3) 

liste = file_to_df("103sr8_CO_q.cmap")
cmap23DB = pa.read_csv("103sr8_CO_q.cmap", sep='\t', skiprows=liste) 
cmap23DB.head()
```
